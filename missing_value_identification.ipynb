```python
import pandas as pd

columns = ['kwh','interval_end','interval_width','site_id']
df = pd.read_csv("vltsdb.csv", names= columns)
df['kw'] = df['kwh']*(3600./df['interval_width'])
df
```

```python
site_ids = df['site_id'].unique()
len(site_ids)
```

```python
hourly_sites = []
fifteen_min_sites = []
misc_sites = []

for i in range(0,len(site_ids)):
    df1= df.loc[df['site_id']== site_ids[i]]
    print(site_ids[i]+' ===>')
    df1.interval_end = pd.to_datetime(df1.interval_end)
    df1.index= df1.interval_end
    df1.sort_index()
    #df1 = df1['kw']
    if df1.shape[0] < 24:
        misc_sites.append(site_ids[i])
        print('yes')
    else:
        status = pd.infer_freq(df1.index)
        if status =='15T':
            fifteen_min_sites.append(site_ids[i])
        if status =='H':
            hourly_sites.append(site_ids[i]) 
        #print(status)
        if status == None:
            df1['minute'] = pd.to_datetime(df1['interval_end']).dt.minute
            sampling_interval = df1.minute.unique()
            checkpoint1 = 15 in sampling_interval
            checkpoint2 = (df1['minute'] == 0).all()
            if checkpoint1 == True:
                fifteen_min_sites.append(site_ids[i]) 
            if checkpoint2 == True:
                hourly_sites.append(site_ids[i])
print(len(fifteen_min_sites), len(hourly_sites), len(misc_sites))
```

```python
import pickle

file_name = "fifteen_minute_sampled.pkl"

open_file = open(file_name, "wb")
pickle.dump(fifteen_min_sites, open_file)
open_file.close()
```

```python
file_name = "hourly_sampled.pkl"

open_file = open(file_name, "wb")
pickle.dump(hourly_sites, open_file)
open_file.close()
```

```python
hourly_missing_values = []
for i in range(0,len(hourly_sites)):
    df1= df.loc[df['site_id']== hourly_sites[i]]
    print(hourly_sites[i]+' ===>')
    df1.interval_end = pd.to_datetime(df1.interval_end)
    df1.index= df1.interval_end
    df1.sort_index()
    df1 = df1['kw']
    a = df1.index[df1.index.to_series().diff().ne(pd.Timedelta('1H'))][1:]
    pos = df1.index.get_indexer(a)
    idx = df1.index.astype(str)
    out = [list(x) for x in zip(idx[pos-1], idx[pos])]
    data = pd.DataFrame({"site_id": [hourly_sites[i]],"missing_data":[out]})
    hourly_missing_values.append(data)
    print (out)
```

```python
df_hourly_missing = pd.concat(hourly_missing_values)
df_hourly_missing
```

```python
df_hourly_missing.to_csv('hourly_sampled_siteid_missing_timestamps.csv')
```

```python
fifteen_min_sites_missing_values = []
for i in range(0,len(fifteen_min_sites)):
    df1= df.loc[df['site_id']== fifteen_min_sites[i]]
    print(fifteen_min_sites[i]+' ===>')
    df1.interval_end = pd.to_datetime(df1.interval_end)
    df1.index= df1.interval_end
    df1.sort_index()
    df1 = df1['kw']
    a = df1.index[df1.index.to_series().diff().ne(pd.Timedelta(minutes=15))][1:]
    pos = df1.index.get_indexer(a)
    idx = df1.index.astype(str)
    out = [list(x) for x in zip(idx[pos-1], idx[pos])]
    data = pd.DataFrame({"site_id": [fifteen_min_sites[i]],"missing_data":[out]})
    fifteen_min_sites_missing_values.append(data)
    print (out)
```

```python
df_fifteen_min_sampled_siteids_missing = pd.concat(fifteen_min_sites_missing_values)
df_fifteen_min_sampled_siteids_missing.to_csv('fifteen_min_sampled_siteids_missing_timestamps.csv')
df_fifteen_min_sampled_siteids_missing
```
